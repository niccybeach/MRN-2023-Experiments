```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python


data = ["C:/Users/nickp/.an MRN experiments/PIVlab/8mm new/PIVlab_Extr_Velocity magnitude [m per s]_1_909.txt",
"C:/Users/nickp/.an MRN experiments/PIVlab/8mm new/PIVlab_Extr_Velocity magnitude [m per s]_2_1392.txt",
"C:/Users/nickp/.an MRN experiments/PIVlab/8mm new/PIVlab_Extr_Velocity magnitude [m per s]_3_1767.txt",
"C:/Users/nickp/.an MRN experiments/PIVlab/8mm new/PIVlab_Extr_Velocity magnitude [m per s]_4_1507_1507.txt",
"C:/Users/nickp/.an MRN experiments/PIVlab/8mm new/PIVlab_Extr_Velocity magnitude [m per s]_5_1730.txt"]
    
new_cols = ['d [m]', 'x [m]', 'y [m]', 'v [m/s]']
d_1 = []
v_mag1 = []

for i in data:
    dr = pd.read_csv(i)
    dr.columns = new_cols
    d_1.append(dr['d [m]'])
    v_mag1.append(dr['v [m/s]'])
#     display(dr)

plt.figure()
plt.title('High drop tests')
plt.xlabel('Distance on line [m]')
plt.ylabel('velocity magnitude [m/s]')
plt.grid()
label = ['8mm bead test 1',
         '8mm bead test 2',
         '8mm bead test 3',
         '8mm bead test 4',
        '8mm bead test 5']

for j in range(0, 5):
    plt.plot(d_1[j], v_mag1[j], label=label[j])
    plt.legend()
plt.show()

max_avg = []
for i in v_mag1:
    max_avg.append(max(i))

print(max_avg)
print(np.average(np.array([max_avg])))
print(np.std(np.array([max_avg])))
```
